```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

# импортируем модули math и numpy
import math
import numpy
# импортируем модуль matplotlib.pyplot и присваиваем ему имя mpp
import matplotlib.pyplot as mpp

# если файл запущен как скрипт
if __name__=='__main__':
    # создаем переменную arguments и присваиваем ей каждое 0.1-ое значение списка numpy.r_ в диапозоне от 0 до 200
    arguments = numpy.r_[0:200:0.1]
    # вызываем функцию plot() модуля matplotlib.pyplot,
    mpp.plot(
        # передавая ей параметр arguments
        arguments,
        # и список, состоящий из значений функции sin(x)/sin(x/20.0), вычисленной для всех элементов списка arguments
        [math.sin(a) * math.sin(a/20.0) for a in arguments]
    )
    # вызываем функцию show() модуля matplotlib.pyplot
    mpp.show()
```
