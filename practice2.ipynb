# Приблизительное вычисление $e^x$

$e^x=\sum_{n=0}^{\infty}\frac{x^n}{n!}$

```python
import math

iterations=50

def e(x):
    n=0
    xn=1
    facn=1
    epsilon=1
    while n<iterations:
        n+=1
        xn*=x
        facn*=n
        epsilon+=xn/facn
    return epsilon

x=1
while x<5:
    print(x,math.exp(x),e(x))
    x+=0.1
```

Как видно, значение функции вычисляется нами очень точно.

```python
import numpy
import matplotlib.pyplot as mpp

arguments = numpy.r_[-5:5:0.1]
#0^0 неопределено, однако значение 0 из-за несоверщенства двоичных вычислений дробей не достигается
mpp.plot(
        arguments,
        [e(a) for a in arguments]
    )
mpp.plot(
        arguments,
        [math.exp(a) for a in arguments]
    )
mpp.show()
```
